```python
import numpy as np
import matplotlib.pyplot as plt

# Multivariable nonlinear function
def f(x):
    return (x[0]**2+5*(x[1]-6)**2+5)

# Gradient
def grad_f(x):
    return np.array([2 * x[0], +10 * (x[1]-6)])

# Gradient Descent Parameters
learning_rate = 0.03
n_iterations = 30
x_start = np.array([3.0, -5.0])

# History for plotting
x_history = [x_start.copy()]
f_history = [f(x_start)]

x = x_start.copy()

for i in range(n_iterations):
    grad = grad_f(x)
    x = x - learning_rate * grad

    x_history.append(x.copy())
    f_history.append(f(x))

    print(f"Iter {i+1}: x = {x}, f(x) = {f(x):.4f}")

```

```python
# Plot in 2D case
if len(x_start) == 2:
    x_vals = [x[0] for x in x_history]
    y_vals = [x[1] for x in x_history]


    X, Y = np.meshgrid(np.linspace(-10, 10, 100), np.linspace(-10, 10, 100))
    Z = (X**2+5*(Y-6)**2+5)

    plt.figure(figsize=(6, 6))
    p1=plt.contour(X, Y, Z, levels=200, cmap='viridis')
    plt.colorbar(p1, label='Loss')
    plt.plot(x_vals, y_vals, marker='o', color='red', label='Descent Path')
    plt.title('Gradient Descent on f(x)')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid(True)
    plt.axis('equal')
    plt.show()
```
